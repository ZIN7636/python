```python
import tkinter as tk
from math import sin, cos, tan, log, sqrt, pi, e, radians

# 메인 윈도우 생성
root = tk.Tk()
root.title("공학용 계산기")
root.geometry("480x500")
root.resizable(False, False)

# 입력창
entry = tk.Entry(root, width=20, font=('Arial', 24), bd=10, relief='sunken', justify='right')
entry.grid(row=0, column=0, columnspan=7, pady=10)

# 버튼 누를 때 실행되는 함수
def button_click(char):
    if char == 'AC':
        entry.delete(0, tk.END)
    elif char == '=':
        try:
            expression = entry.get().replace('÷', '/').replace('×', '*').replace('√', 'sqrt')\
                                    .replace('π', str(pi)).replace('e', str(e)).replace('^', '**')
            result = eval(expression)
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif char == 'sin':
        try:
            result = sin(radians(float(entry.get())))
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif char == 'cos':
        try:
            result = cos(radians(float(entry.get())))
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif char == 'tan':
        try:
            result = tan(radians(float(entry.get())))
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif char == 'log':
        try:
            result = log(float(entry.get()), 10)
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif char == 'ln':
        try:
            result = log(float(entry.get()))
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif char == '√':
        try:
            result = sqrt(float(entry.get()))
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    else:
        entry.insert(tk.END, char)

# 버튼 목록 (7칸씩 맞춰서 이미지처럼 구성)
buttons = [
    ['Rad', 'Deg', 'x!', '(', ')', '%', 'AC'],
    ['Inv', 'sin', 'ln', '7', '8', '9', '÷'],
    ['π', 'cos', 'log', '4', '5', '6', '×'],
    ['e', 'tan', '√', '1', '2', '3', '−'],
    ['Ans', 'EXP', '^', '0', '.', '=', '+']
]

# 버튼 생성
for i, row in enumerate(buttons):
    for j, char in enumerate(row):
        btn = tk.Button(root, text=char, width=6, height=2, font=('Arial', 14),
                        command=lambda c=char: button_click(c))
        btn.grid(row=i+1, column=j, padx=3, pady=3)

root.mainloop()
```
